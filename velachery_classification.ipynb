```python
from sklearn.ensemble import  RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier

from sklearn import preprocessing #For scaling

from sklearn.preprocessing import LabelEncoder # Converting Categorical(text) to Categorical(numerical)
```

```python
q=[]
pol=['PM2.5','CO','SO2','NO2']
x1=[]
x2=[]
x3=[]
x4=[]
xmax=[]
xpol=[]



for i in range(0,df.shape[0]):
    B=[]
    a=df.iloc[i]['PM2.5']
    if a<=30 and a>=0:
        b=1
    elif a<=60 and a>=31:
        b=2
    elif a<=90 and a>=61:
        b=3
    elif a<=120 and a>=91: 
        b=4
    elif a<=250 and a>=121:
        b=5
    elif a>251:
        b=6
    x1.append(b)
    B.append(b)
        
    a=df.iloc[i]['CO']
    a/=1000
    if a<=1.0 and a>=0:
        b=1
    elif a<=2.0 and a>=1.1:
        b=2
    elif a<=10 and a>=2.1:
        b=3
    elif a<=10.1 and a>=17: 
        b=4
    elif a<=34 and a>=17.1:
        b=5
    elif a>34:
        b=6
    x2.append(b)
    B.append(b)
    
    a=df.iloc[i]['SO2']
    if a<=40 and a>=0:
        b=1
    elif a<=80 and a>=41:
        b=2
    elif a<=380 and a>=81:
        b=3
    elif a<=800 and a>=381: 
        b=4
    elif a<=1600 and a>=801:
        b=5
    elif a>1600:
        b=6
    x3.append(b)        
    B.append(b)
    
    a=df.iloc[i]['NO2']
    if a<=40 and a>=0:
        b=1
    elif a<=80 and a>=41:
        b=2
    elif a<=180 and a>=81:
        b=3
    elif a<=280 and a>=181: 
        b=4
    elif a<=400 and a>=281:
        b=5
    elif a>400:
        b=6
    x4.append(b)
    B.append(b)
    
    
    qu=np.array(B)
    mq=qu.max()
    xmax.append(mq)
    po=pol[np.argmax(qu)]
    xpol.append(po)
    #print(B,mq,po)
    
    
    q.append(mq)
```

```python
q2=pd.DataFrame(q[:])
q2[0].value_counts()
```

```python
quality=pd.DataFrame({'qPM2.5':np.array(x1),'qCO':np.array(x2),'qSO2':np.array(x3),
                      'qNO2':np.array(x4),
                      'max_quality':np.array(xmax),'primary pollutant':np.array(xpol)})
quality.head()  
```

```python
quality.to_csv(r'quality labels for each pollutant.csv',index=None,header=True)
```

```python
x = df.drop(['Quality','Primary Pollutant','Year'], 1)  
y =  df['Quality']
x.shape
x
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y,test_size=0.3, random_state=0)  
```

```python
from sklearn.metrics import confusion_matrix  
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
lr = LogisticRegression()
lr.fit(xtrain,ytrain)
ypred=lr.predict(xtest)
cm= confusion_matrix(ytest, ypred)  
print(cm)  
print('Accuracy : '+ str(accuracy_score(ytest, ypred)))
print(classification_report(ytest,ypred))
```

```python
model=DecisionTreeClassifier()
model.fit(xtrain,ytrain)
ypred = model.predict(xtest)
cm= confusion_matrix(ytest, ypred)  
print(cm)  
print('Accuracy : '+ str(accuracy_score(ytest, ypred)))
print(classification_report(ytest,ypred))
```

```python
model = GaussianNB()
model.fit(xtrain, ytrain)
ypred = model.predict(xtest)
cm= confusion_matrix(ytest, ypred)  
print(cm)  
print('Accuracy : ' +str(accuracy_score(ytest, ypred)))
print(classification_report(ytest,ypred))
```

```python
classifier=RandomForestClassifier()
classifier.fit(xtrain,ytrain)
ypred = classifier.predict(xtest) 
cm = confusion_matrix(ytest, ypred)  
print(cm)  
print('Accuracy : '+ str(accuracy_score(ytest, ypred)))
print(classification_report(ytest,ypred))
```

```python
knn = KNeighborsClassifier()
knn.fit(xtrain, ytrain)
ypred = knn.predict(xtest)
cm= confusion_matrix(ytest, ypred)  
print(cm)  
print('Accuracy : '+ str(accuracy_score(ytest, ypred)))
print(classification_report(ytest,ypred))
```

```python
RF=RandomForestClassifier()
RF_boost=AdaBoostClassifier(base_estimator=RF,n_estimators=50)
RF_boost.fit(xtrain, ytrain)
ypred = RF_boost.predict(xtest)
cm= confusion_matrix(ytest, ypred)  
print(cm)  
print('Accuracy : '+ str(accuracy_score(ytest, ypred)))
print(classification_report(ytest,ypred))
```

```python
from sklearn.model_selection import GridSearchCV
RF=RandomForestClassifier()
params={'max_depth':np.arange(1,20),'criterion':['entropy','gini']}
GS=GridSearchCV(RF,params,cv=5,scoring='f1_micro')
GS.fit(x,y)
GS.best_params_
```

```python
RF_boost=AdaBoostClassifier(base_estimator=RFt,n_estimators=50)
RF_boost.fit(xtrain, ytrain)
ypred = RF_boost.predict(xtest)
cm= confusion_matrix(ytest, ypred)  
print(cm)  
print('Accuracy : '+ str(accuracy_score(ytest, ypred)))
print(classification_report(ytest,ypred))
```
